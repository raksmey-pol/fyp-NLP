## 5. Load and Test Best Model

```python
# Load the best model for interactive testing
best_model_name = df_results.iloc[0]['Model'].lower()

if best_model_name in ['lstm', 'bilstm', 'cnn_lstm']:
    # Load deep learning model
    from src.models.deep_learning_models import LSTMModel, BiLSTMModel, CNNLSTMModel
    
    ModelClass = {'lstm': LSTMModel, 'bilstm': BiLSTMModel, 'cnn_lstm': CNNLSTMModel}[best_model_name]
    model = ModelClass(max_features=10000, embedding_dim=128, max_length=500)
    model.load(best_model_name)
    
    def predict_news(text):
        X = model.prepare_data([text])
        proba = model.predict_proba(X)[0]
        prediction = 'REAL' if proba >= 0.5 else 'FAKE'
        confidence = proba if proba >= 0.5 else 1 - proba
        return prediction, confidence
else:
    # Load traditional ML model
    from src.models.traditional_models import TraditionalModelTrainer
    
    model = TraditionalModelTrainer(model_type=best_model_name)
    model.load(f'{best_model_name}_model.pkl')
    
    # Load TF-IDF vectorizer for preprocessing
    import pickle
    with open('../data/processed/features/tfidf/tfidf_vectorizer.pkl', 'rb') as f:
        vectorizer = pickle.load(f)
    
    def predict_news(text):
        X = vectorizer.transform([text])
        proba = model.model.predict_proba(X)[0, 1]
        prediction = 'REAL' if proba >= 0.5 else 'FAKE'
        confidence = proba if proba >= 0.5 else 1 - proba
        return prediction, confidence

print(f"✅ Loaded best model: {best_model_name.upper()}")
print("Ready for predictions!")
```

## 6. Interactive Prediction Testing

```python
# Test with sample texts
test_texts = [
    "Scientists discover new planet in habitable zone of nearby star system.",
    "BREAKING: Aliens land in New York City, demand to speak with president!",
    "Stock market reaches record high amid strong economic recovery.",
    "You won't believe what this celebrity did! Doctors hate them!"
]

print("\n" + "="*80)
print("PREDICTION EXAMPLES")
print("="*80)

for text in test_texts:
    prediction, confidence = predict_news(text)
    print(f"\nText: {text}")
    print(f"Prediction: {prediction} (Confidence: {confidence:.2%})")
    print("-" * 80)
```
